```python
from lib.stats_aplt import Df_Stats

houses_url = 'https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv'
houses = Df_Stats(houses_url)

assert houses is not None, 'Error creating Df_Stats object'
assert houses.df.shape == (506, 14), 'Dataframe shape is incorrect'
```

```python
import inspect
# Get a list of methods using inspect module
methods_list = [method[0] for method in inspect.getmembers(
    houses, predicate=inspect.ismethod)]
 
print("Methods using inspect module:", methods_list)
```

```python
columns = ['crim', 'zn', 'indus', 'chas',  'rm', 'age']
for col in columns:
    mean_test = houses.variable_mean(col)
    quantile_test = houses.variable_percentile(col, 25)
    median_test = houses.variable_median(col)
    std_test = houses.variable_std(col)
    print(std_test)
    describe_test = houses.stats_summary([col])
    assert describe_test.loc["mean", col] == mean_test
    assert describe_test.loc["std", col] == std_test
    assert describe_test.loc["25%", col] == quantile_test
```

```python
for c in columns:
    houses.plot_hist_var(c, True, c)
```

```python
for c in columns:
    for c2 in columns:
        if c != c2:
            houses.plot_scatter_two_vars(c, c2, True, c)
```
